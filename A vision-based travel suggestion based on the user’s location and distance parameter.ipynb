```python
import requests
import pandas as pd
from pandas import option_context
from collections import defaultdict

url = 'https://api.geoapify.com/v2/places'

category = input("Enter a category from the list available : ")
params = dict(categories = category, filter='circle:%s,%s,40000' % (float(ls[1]), float(ls[0])), limit= 20, apiKey='5b73fb5c534b45fb8673975a0c905347')

resp = requests.get(url=url, params=params)
data = resp.json()
if ('error' in data):
    print("Enter a Valid Category")
else:
    n = []
    c = []
    s = []
    lon = []
    lat = []
    add = []
    for i in data['features']:
        if ('name' not in i['properties']):
            n.append("Not Available")
        else:
            n.append(i['properties']['name'])

        if ('city' not in i['properties']):
            c.append("Not Available")
        else:
            c.append(i['properties']['city'])

        if ('state' not in i['properties']):
            s.append("Not Available")
        else:
            s.append(i['properties']['state'])

        if ('lon' not in i['properties']):
            lon.append(0)
        else:
            lon.append(i['properties']['lon'])

        if ('lat' not in i['properties']):
            lat.append(0)
        else:
            lat.append(i['properties']['lat'])

        if ('address_line2' not in i['properties']):
            add.append("Not Available")
        else:
            add.append(i['properties']['address_line2'])

    d = list(zip(n, c, s, lon, lat, add))

    ls1 = [[i[0], i[1], i[2], i[5]] for i in d]
    col = ['Name', 'City', 'State', 'Address']
    df = pd.DataFrame(ls1, columns = col)
    with option_context('display.max_colwidth', 800):
        display(df)
```

```python
import math

def haversine(lat1, lon1, lat2, lon2):
    dLat = (lat2 - lat1) * math.pi / 180.0
    dLon = (lon2 - lon1) * math.pi / 180.0
    
    lat1 = (lat1) * math.pi / 180.0
    lat2 = (lat2) * math.pi / 180.0
    
    a = (pow(math.sin(dLat / 2), 2) + pow(math.sin(dLon / 2), 2) * math.cos(lat1) * math.cos(lat2))
    rad = 6371
    c = 2 * math.asin(math.sqrt(a))
    return rad * c


lat1 = float(ls[1])
lon1 = float(ls[0])
for i in range(len(d)):
    lat2 = d[i][3]
    lon2 = d[i][4]
    d[i] = list(d[i])
    d[i].append(round(haversine(lat1, lon1,lat2, lon2), 2)), "K.M"
```
